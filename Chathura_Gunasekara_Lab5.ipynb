```python
import urllib2
myurl = 'http://www.city-data.com/top1.html'
sock = urllib2.urlopen(myurl) 
lines = sock.readlines() 
sock.close()
```

```python
import re
lineNum=0
print "Rank\t","City\t\t","Population\t\t\t","Link"
for line in lines:
    if "href='/city/" in line:
        city = line.split('>')[9].split("</a")[0]
        pop = line.split('>')[10].split(')')[0].split(' ')[2]
        repl1 = re.sub('[,]', '', city)
        url = re.sub('[ ]','-',repl1).strip()
        lineNum = lineNum + 1
        print lineNum,"\t",city," ",pop,"\thttp://www.city-data.com/city/"+url+".html"
        print "-------------------------------------------------------------------------------------------"
        print ""
        #print line.split('>')[9].split("</a")[0],line.split('>')[10].split(')')[0].split(' ')[2]
    if "javascript" in line:
        break
```
